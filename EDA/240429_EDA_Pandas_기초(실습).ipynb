```python
#01. 여성 승객들의 평균 나이를 계산해보세요.
np.mean(titanic.loc[titanic.Sex == 'female', 'Age'])
```

```python
titanic.loc[titanic.Sex == 'female', 'Age'].mean()
```

```python
# 02. 1등석에 탑승한 승객 중 최대 요금을 낸 사람의 이름을 찾아주세요.
# 1등석에 탑승한 승객 >

# titanic.loc[titanic.Pclass == 1, 'Fare'].max()
# titanic.loc[titanic.Pclass == 1, ['Name', 'Fare']]
temp = titanic.loc[titanic.Pclass == 1, ['Name', 'Fare']]
# temp[temp.Fare == temp.Fare.max()]
temp.loc[temp.Fare == temp.Fare.max(), "Name"]
```

```python
# 03. 1등석에 탑승한 승객 중 승선한 곳이 Queenstoen인 사람들의 수는?
temp2 = titanic[titanic.Pclass == 1]
temp2[temp2.Embarked == 'Q']
len(temp2[temp2.Embarked == 'Q'])
```

```python
# mask & mask : 두 조건을 다 만족하는 mask (AND)
# mask | mask : 두 조건중 하나 이상 만족하는 mask (OR)
# ~ mask : 조건을 반전 (NOT 연산)
len(titanic[(titanic.Pclass == 1) & (titanic.Embarked == 'Q')])
```

```python
# 04. 승선한 곳이 "S"인 사람들의 생존자 수는?
# len(titanic[(titanic.Embarked == 'S') & (titanic.Survived == 1)])
titanic.loc[titanic.Embarked.isin(['S']), 'Survived'].sum()
```

```python
# Q5. 미혼 여성중에 나이를 모르는 사람의 수는?
titanic.loc[titanic.Name.str.contains('Miss.'), 'Age'].isnull().sum()
```
