```python
import scipy
from sklearn import metrics

# sse 함수: 표준 제곱 오차(Standard Squared Error) 계산
    # clf: 선형 회귀 모델 (예: LinearRegression 객체)
    # X: 입력 데이터 (피처)
    # y: 실제 타겟 값
    # y_hat: 모델이 예측한 타겟 값
    # sse: 표준 제곱 오차
    # 반환 값: 모델의 표준 제곱 오차
def sse(clf, X, y):
    y_hat = clf.predict(X)
    sse = np.sum((y_hat - y) ** 2)
    return sse / X.shape[0]

# adj_r2_score 함수: 조정 R^2 계산
    # n: 데이터 샘플 수
    # p: 피처 수
    # r_squared: 모델의 R^2 점수
    # 반환 값: 모델의 조정된 R^2 점수
def adj_r2_score(clf, X, y):
    n = X.shape[0]  # 관측치 수 Number of observations
    p = X.shape[1]  # 피쳐 수 Number of features
    r_squared = metrics.r2_score(y, clf.predict(X))
    return 1 - (1 - r_squared) * ((n - 1) / (n - p - 1))

# coef_se 함수: 회귀 계수의 표준 오차(Standard Error) 계산
    # X1: 입력 데이터에 상수(intercept) 항 추가
    # se_matrix: 표준 오차 행렬
    # 반환 값: 회귀 계수의 표준 오차 배열
def coef_se(clf, X, y):
    n = X.shape[0]
    X1 = np.hstack((np.ones((n, 1)), np.matrix(X)))
    se_matrix = scipy.linalg.sqrtm(
        metrics.mean_squared_error(y, clf.predict(X)) *
        np.linalg.inv(X1.T * X1)
    )
    return np.diagonal(se_matrix)

# coef_tval 함수: 회귀 계수의 t-통계량(t-statistic) 계산
    # a: 절편의 t-통계량
    # b: 회귀 계수의 t-통계량
    # 반환 값: t-통계량 배열
def coef_tval(clf, X, y):
    a = np.array(clf.intercept_ / coef_se(clf, X, y)[0])
    b = np.array(clf.coef_ / coef_se(clf, X, y)[1:])
    return np.append(a, b)

# coef_pval 함수: 회귀 계수의 p-값 계산
    # t: t-통계량 배열
    # p: p-값 배열
    # 반환 값: p-값 배열
def coef_pval(clf, X, y):
    import scipy.stats
    n = X.shape[0]
    t = coef_tval(clf, X, y)
    p = 2 * (1 - scipy.stats.t.cdf(np.abs(t), n - 1))
    return p

# summary 함수: 회귀 모델 요약 통계 출력
    # xlabels: 피처의 레이블 (없을 경우 기본 레이블 생성)
    # coef_df: 회귀 계수 요약 데이터프레임
    # 모델의 R^2, 조정 R^2, MSE와 함께 회귀 계수, 표준 오차, t-통계량, p-값을 출력
def summary(clf, X, y, xlabels=None):
    # Check and/or make xlabels
    ncols = X.shape[1]
    if xlabels is None:
        xlabels = np.array(
            ['x{0}'.format(i) for i in range(1, ncols + 1)], dtype='str')
    elif isinstance(xlabels, (tuple, list)):
        xlabels = np.array(xlabels, dtype='str')
    # Make sure dims of xlabels matches dims of X
    if xlabels.shape[0] != ncols:
        raise AssertionError(
            "Dimension of xlabels {0} does not match "
            "X {1}.".format(xlabels.shape, X.shape))
    # Create data frame of coefficient estimates and associated stats
    coef_df = pd.DataFrame(
        index=['_intercept'] + list(xlabels),
        columns=['Estimate', 'Std. Error', 't value', 'p value']
    )
    try:
        coef_df['Estimate'] = np.concatenate(
            (np.round(np.array([clf.intercept_]), 6), np.round((clf.coef_), 6)))
    except Exception as e:
        coef_df['Estimate'] = np.concatenate(
            (
                np.round(np.array([clf.intercept_]), 6),
                np.round((clf.coef_), 6)
            ), axis = 1
    )[0,:]
    coef_df['Std. Error'] = np.round(coef_se(clf, X, y), 6)
    coef_df['t value'] = np.round(coef_tval(clf, X, y), 4)
    coef_df['p value'] = np.round(coef_pval(clf, X, y), 6)
    # Output results
    print('Coefficients:')
    print(coef_df.to_string(index=True))
    print('---')
    print('R-squared:  {0:.6f},    Adjusted R-squared:  {1:.6f},    MSE: {2:.1f}'.format(
        metrics.r2_score(y, clf.predict(X)), adj_r2_score(clf, X, y), sse(clf, X, y)))
```

- MinMaxScaler(): MinMaxScaler 객체를 생성합니다. 이 객체는 데이터를 [0, 1] 범위로 스케일링하는 데 사용됩니다.
- .fit(X.iloc[train_idx]): 학습 데이터 (X.iloc[train_idx])를 사용하여 스케일러를 학습시킵니다. 스케일러는 각 피처의 최소값과 최대값을 학습하여 변환에 사용합니다.
- scaler.transform(X): 전체 데이터 (X)를 학습된 스케일러를 사용하여 변환합니다. 각 피처의 값은 [0, 1] 범위로 조정됩니다.
- X_scal 변수에 변환된 데이터를 저장합니다.
- pd.DataFrame(X_scal, columns=X.columns): 변환된 데이터를 원래 데이터의 열 이름을 사용하여 데이터프레임으로 변환합니다.

[전체 과정 요약]
1. MinMaxScaler 객체를 생성하고, 학습 데이터로 스케일러를 학습시킵니다.
2. 학습된 스케일러를 사용하여 전체 데이터를 변환합니다.
3. 변환된 데이터를 데이터프레임 형태로 변환하여 원래 데이터의 열 이름을 유지합니다.

[Ridge Regression]
 - Hyperparameter Tuning using for Loop
 - Hyperparameter Tuning using GridSearchCV

[Ridge Regression Parameters]
   - Package : https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html
   - alpha : L2-norm Penalty Term
     - alpha : 0 일 때, Just Linear Regression
   - fit_intercept : Centering to zero
     - 베타0를 0로 보내는 것 (베타0는 상수이기 때문에)
   - max_iter : Maximum number of interation
     - Loss Function의 Ridge Penalty Term은 Closed Form 값이기는 하지만 값을 찾아 나감
     - Penalty Term : (1 / (2 * n_samples)) * ||y - Xw||^2_2 + alpha * ||w||_2



---
이 코드는 주어진 alpha 값들에 대해 Ridge 회귀 모델을 학습시키고, 각 모델의 R^2, MSE, RMSE를 계산하여 출력합니다. 이를 통해 최적의 alpha 값을 선택할 수 있습니다.

Ridge 회귀 모델을 다양한 alpha 값으로 학습시키고, 각 모델의 성능을 평가하여 최적의 alpha 값을 선택하는 과정을 보여줍니다.

- penelty 리스트는 Ridge 회귀 모델의 alpha 값 후보들을 나타냅니다. alpha 값은 정규화의 강도를 조절하는 하이퍼파라미터입니다.

- Ridge(alpha=a): alpha 값을 사용하여 Ridge 회귀 모델을 생성합니다.
- .fit(X_scal.iloc[train_idx], Y.iloc[train_idx]): 학습 데이터
- (X_scal.iloc[train_idx], Y.iloc[train_idx])를 사용하여 모델을 학습시킵니다

- model.score: 검증 데이터(X_scal.iloc[valid_idx], Y.iloc[valid_idx])를 사용하여 모델의 R^2 점수를 계산

- model.predict: 검증 데이터(X_scal.iloc[valid_idx])를 사용하여 타겟 값을 예측.
- mean_squared_error: 실제 타겟 값(Y.iloc[valid_idx])과 예측된 값(pred_y) 사이의 평균 제곱 오차(MSE)를 계산


이 코드를 실행하면, 다양한 alpha 값에 대해 Ridge 회귀 모델의 성능을 평가한 결과를 확인할 수 있습니다. 이를 통해 가장 성능이 좋은 alpha 값을 선택할 수 있습니다.

```python
penelty = [0.00001, 0.00005, 0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.6, 0.7, 0.9, 1, 10]

for a in penelty:
    model = Ridge(alpha=a).fit(X_scal.iloc[train_idx], Y.iloc[train_idx])
    score = model.score(X_scal.iloc[valid_idx], Y.iloc[valid_idx])
    pred_y = model.predict(X_scal.iloc[valid_idx])
    mse = mean_squared_error(Y.iloc[valid_idx], pred_y)
    print("Alpha:{0:5f}, R2:{1:.7f}, MSE:{2:.7f}, RMSE:{3:.7f}".format(a, score, mse, np.sqrt(mse)))
```

```python
penelty = [0.00001, 0.00005, 0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.6, 0.7, 0.9, 1, 10]

# Using For Loop !!
# Ridge Regression
# select alpha by checking R2, MSE, RMSE
for a in penelty:
    model = Ridge(alpha=a).fit(X_scal.iloc[train_idx], Y.iloc[train_idx])
    score = model.score(X_scal.iloc[valid_idx], Y.iloc[valid_idx])
    pred_y = model.predict(X_scal.iloc[valid_idx])
    mse = mean_squared_error(Y.iloc[valid_idx], pred_y)
    # 각 alpha 값에 대해 R^2 점수, MSE, RMSE(루트 평균 제곱 오차)를 출력
    print("Alpha:{0:.5f}, R2:{1:.7f}, MSE:{2:.7f}, RMSE:{3:.7f}".format(a, score, mse, np.sqrt(mse)))
```



---

이 코드 조각은 최적의 alpha 값을 사용하여 Ridge 회귀 모델을 학습시키고, 검증 데이터에 대해 모델의 요약 통계를 출력하는 과정입니다.

- Ridge(alpha=0.01): alpha 값을 0.01로 설정하여 Ridge 회귀 모델을 생성합니다. 이는 이전에 다양한 alpha 값들에 대해 성능을 비교한 후 선택된 최적의 값입니다.
- .fit(X_scal.iloc[train_idx], Y.iloc[train_idx]): 학습 데이터 (X_scal.iloc[train_idx], Y.iloc[train_idx])를 사용하여 모델을 학습.
- model_best 변수에 학습된 모델을 저장.

- summary 함수는 학습된 모델 model_best, 검증 데이터 X_scal.iloc[valid_idx] (검증 데이터의 입력 변수), Y.iloc[valid_idx] (검증 데이터의 타겟 변수), 및 입력 변수의 열 이름 X_scal.columns을 인자로 받습니다.
- 이 함수는 모델의 회귀 계수, 표준 오차, t-값, p-값, R^2, 조정된 R^2 및 MSE를 계산하여 출력합니다.

이 코드를 실행하면, 최적의 alpha 값으로 학습된 Ridge 회귀 모델의 검증 데이터에 대한 요약 통계가 출력됩니다. 이 요약 통계는 모델의 성능을 평가하고, 회귀 계수의 유의성을 파악하는 데 도움이 됩니다.

```python
model_best = Ridge(alpha=0.01).fit(X_scal.iloc[train_idx], Y.iloc[train_idx])
summary(model_best, X_scal.iloc[valid_idx], Y.iloc[valid_idx], xlabels = X_scal.columns)
```



---

RidgeCV를 사용하여 최적의 alpha 값을 찾고, 그 값으로 Ridge 회귀 모델을 학습시키는 과정을 보여줍니다. GridSearchCV와 유사하게 작동하지만, Ridge 회귀에 특화된 RidgeCV를 사용합니다.

- RidgeCV는 Ridge 회귀의 교차 검증을 수행하여 최적의 alpha 값을 찾는 클래스입니다.
- alphas=penelty는 alpha 값의 후보 리스트를 설정합니다. 여기서 penelty 리스트는 다양한 alpha 값들을 포함하고 있습니다.
- cv=5는 5-폴드 교차 검증을 사용하여 모델을 평가합니다.

```python
# Using GridSearchCV
ridge_cv=RidgeCV(alphas=penelty, cv=5)
model = ridge_cv.fit(X_scal.iloc[train_idx], Y.iloc[train_idx])
print("Best Alpha:{0:.5f}, R2:{1:.4f}".format(model.alpha_, model.best_score_))
```



---
RidgeCV로 최적의 alpha 값을 찾은 후, 해당 alpha 값을 사용하여 Ridge 회귀 모델을 학습시키고, 검증 데이터에 대해 모델의 성능을 평가하며, 최종적으로 모델의 요약 통계를 출력하는 과정을 보여줍니다.

- Ridge(alpha=model.alpha_)는 RidgeCV로 찾은 최적의 alpha 값을 사용하여 Ridge 회귀 모델을 생성합니다.
- .fit(X_scal.iloc[train_idx], Y.iloc[train_idx])는 학습 데이터 (X_scal.iloc[train_idx], Y.iloc[train_idx])를 사용하여 모델을 학습시킵니다.
- model_best 변수에 학습된 모델을 저장합니다.
- score = model_best.score(X_scal.iloc[valid_idx], Y.iloc[valid_idx]): 검증 데이터 (X_scal.iloc[valid_idx], Y.iloc[valid_idx])를 사용하여 모델의 R^2 점수를 계산합니다.
- pred_y = model_best.predict(X_scal.iloc[valid_idx]): 검증 데이터의 입력 변수를 사용하여 타겟 값을 예측합니다.
- mse = mean_squared_error(Y.iloc[valid_idx], pred_y): 실제 타겟 값 (Y.iloc[valid_idx])과 예측된 값 (pred_y) 사이의 평균 제곱 오차(MSE)를 계산합니다.
rmse = np.sqrt(mse): MSE의 제곱근인 RMSE(루트 평균 제곱 오차)를 계산합니다.

이 코드를 실행하면 최적의 alpha 값을 사용하여 학습된 Ridge 회귀 모델의 검증 데이터에 대한 성능 평가와 요약 통계가 출력됩니다. 이를 통해 모델의 성능과 회귀 계수의 유의성을 파악할 수 있습니다.

```python
# GridSearchCV Result
model_best = Ridge(alpha=model.alpha_).fit(X_scal.iloc[train_idx], Y.iloc[train_idx])
score = model_best.score(X_scal.iloc[valid_idx], Y.iloc[valid_idx])
pred_y = model_best.predict(X_scal.iloc[valid_idx])
mse = np.sqrt(mean_squared_error(Y.iloc[valid_idx], pred_y))
print("Alpha:{0:.5f}, R2:{1:.7f}, MSE:{2:.7f}, RMSE:{3:.7f}".format(0.01, score, mse, np.sqrt(mse)))
summary(model_best, X_scal.iloc[valid_idx], Y.iloc[valid_idx], xlabels=X_scal.columns)
```
